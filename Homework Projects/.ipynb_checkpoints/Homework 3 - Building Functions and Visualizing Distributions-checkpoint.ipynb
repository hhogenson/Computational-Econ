7. Create a function that plots 2 to 3 different variables from the dataframe. The function may look something like "def plot_data(df,names, etc...). Each plot should include a legend and title.


```python
def plot_data(df, lstDataColumns):
    num_vars = len(lstDataColumns)
    fig, axs = plt.subplots(1, num_vars, figsize = (20, 6))
    for i in range(num_vars):
        ax = axs[i]
        key = lstDataColumns[i]
        color = "C" + str(i)
        df[key].plot.line(color = color, ax = ax)
        ax.set_title(key, fontsize = 20)
        ax.legend(key)
```

```python
plot_data(df4, ["Chickens", "Eggs"])
```
