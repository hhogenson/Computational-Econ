```python
# create some artificial data
macro_dict = {"GDP":{},
             "Money": {},
             "Real GDP": {},
             "Price Level": {}}
for key in macro_dict.keys():
    for i in range(1990, 2021):
        macro_dict[key][i] = np.random.random() * 10000
macro_dict
```

```python
#make it a dataframe
macro_DF = pd.DataFrame(macro_dict)
macro_DF
```

```python
macro_DF[["GDP"]]
```

```python
#generate velocity
macro_DF["Velocity"] = macro_DF["Price Level"].mul(macro_DF["Real GDP"]).div(macro_DF["Money"])
macro_DF
```

# matplotlib

```python
import matplotlib.pyplot as plt
macro_DF.plot()
```

```python
#customizing plot
fig, ax = plt.subplots(figsize = (20,12))
macro_DF.plot(linewidth = 5, alpha = .7, ax = ax)
```

```python
#splitting data into multiple plots
num_vars = len(macro_DF.keys())
fig, axs = plt.subplots(1, num_vars, figsize = (20, 6))

for i in range(num_vars):
    ax = axs[i]
    key = list(macro_DF.keys())[i]
    color = "C" + str(i)
    macro_DF[key].plot.line(color = color, ax = ax)
    ax.set_title(key, fontsize = 20)
```

## Plotting Income Distribution

```python
#get income data from the government
income_data = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])
income_data = income_data.rename(columns = {"Unnamed: 0": "Household Income"})
income_data.head()
```

```python
income_data.iloc[1:-1]
```

```python
#save series of persons in each income level
income = income_data["Number"][1:-1].div(1000)
income.iloc[0] = 0
income.head()
```

```python
#recreating row indices
income_levels = list(range(0, 201, 5))
income_levels.append(250)
income_levels = [str(val) + "k" for val in income_levels]
income_levels.append("250k+")
income_levels
```

```python
#formatting and cleaning data
income_data = pd.DataFrame({"Income Level":np.array(income_levels),
                           "Households in Bracket":income})
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()
income_data.set_index("Income Level", inplace = True)
income_data
```

```python
income_data.plot.line(secondary_y="Cumulative")
```

```python
#generate % in each bracket
    #get total number of households
num_households = income_data["Households in Bracket"].sum().round(3)
    #% in bracket is equal to number in bracket divided by total number
income_data["% in Bracket"] = income_data["Households in Bracket"].div(num_households)
    #generate cumulative percent
income_data["Cumulative %"] = income_data["% in Bracket"].cumsum().round(3)
income_data
```

```python
#rcParams sets parameters globally
plt.rcParams.update({"font.size": 16})
plot_df = income_data[["% in Bracket", "Cumulative %"]]
fig, ax = plt.subplots(figsize = (15, 7))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .95,
                 alpha = .75,
                 ax = ax)
x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y*100) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
```

```python
# call xlabels
#ax.get_xticklabels()
plot_df.index
```

```python
#add $ to the indices
#new_index= []
#for ix in plot_df.index:
 #   new_index.apped("$" + ix)
new_index = ["$" + str(x) for x in plot_df.index]
new_index
```

```python
# build a dist of income w/ 25k bins
    #start by counting by 5s
income_25k_bin = income_data.loc[::5]
    # append missing final bin
income_25k_bin.loc["250k+"] = income_data.iloc[-1]
income_25k_bin
```

```python
#redefine marginal bins by taking the difference of the cumulative bins
income_25k_bin["Households in Bracket"] = income_25k_bin[
    "Cumulative"].diff().fillna(0)
#regen. % in bracket w/ new bins
income_25k_bin["% in Bracket"] = income_25k_bin["Households in Bracket"].div(num_households).mul(100)
#regen. Cumulative % w/ new bins
income_25k_bin["Cumulative %"] = income_25k_bin["% in Bracket"].cumsum().round(3)
income_25k_bin
```

```python
plt.rcParams.update({"font.size": 20})
plot_df = income_25k_bin[["% in Bracket", "Cumulative %"]].iloc[1:]
fig, ax = plt.subplots(figsize = (15, 10))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .99,
                 alpha = .75,
                 ax = ax)
x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
ax.set_title("Income Distribution in the United States")
```
