Machine learning algorithms have been increasingly used to help predict stock prices and returns both in academia, professional practice, and day trading. Fisher and Krauss (2017) tested the ability of a long short-term memory (LSTM) network to predict price changes for S&P 500’s constituent stocks against that of a random forest, deep neural net, and a logistic regression. Overall, they found that the LSTM outperformed the comparison models except during one key time period: the Great Recession. During the financial crisis, the random forest model not only outperformed the LSTM network but also outperformed its previous returns throughout the decade. This is likely due to the structure of a random forest model. The random forest model utilizes the majority decision of an ensemble of decision trees to make a final prediction. Each of these decision trees is trained on a random subset of the data which reduces the propensity of an algorithm to overfit to the data. This allows the model to filter out noise and outliers, two phenomena found in abundance during volatile recessions. Fisher and Krauss’s data covers the time period from 1992 until 2015, and another dire financial time has occurred since then. This paper will investigate if the random forest model’s ability to remain accurate during volatile times extends to the COVID-19 pandemic. It is my hypothesis that the random forest will not have increased predictive power due to the widespread availability of machine learning algorithms having reduced their ability to accurately predict stock movements as the market has begun to account for their use.

As touched on previously, a random forest model works by developing several decision trees trained on random subsets of the data. Each tree not only has access to a random subset of observation but also a limited number of variables. This is typically the square root of the total number of variables present. The final decision of the model is the decision held by the majority of the decision trees. To predict the outcomes of new data, the new variable must be run through each decision tree and the final “vote” taken. This kind of ensemble method results in a model that is robust to noise and outliers in the data and is less likely to overfit the data. Overfitting is when a model is overly specialized in the training data and, as a result, performs poorly on new data given. The benefits of this kind of robustness are especially obvious when thought of in conjunction with something like a recession or pandemic.

While Fisher & Krauss (2017) primarily focused on the LSTM model and the inclusion of such a model would provide a useful test as to whether or not machine learning algorithms in general experienced a renaissance of predictive power during the COVID-19 pandemic (which is unlikely given their even broader availability and increased attention during the time), such a model will not be included. Rather, a simple increase in the returns generated by the random forest model will be sufficient to validate my hypothesis. 

It is appropriate to use the same structure as used in Fisher and Krauss (2017) so as to reduce the chance that the results are subject to any changes from differences in model structure. As such, the random forest model will consist of 1,000 decision trees each with a maximum depth of 20. Additionally, only the square root of the set of features will be available to each tree. 

Fisher & Kramer’s models rely on the constituent stocks in the S&P 500 index, and the use of many stocks to train the models provides more useful information and may allow the algorithm to make better predictions as it can account for the influence that stocks hold over each other. Training the model on a more limited data set (such as a single stock or a group of related ones), however, may allow the model to learn the finer details of that subset rather than broader generalizations from following multiple stocks, but this approach also risks losing the information and wider variability of training data provided by the previous option. One could circumvent the loss of information from such a narrow focus by providing the price information of a select handful of related stocks (such as those in the same industry or in industries that the main industry relies on). Ultimately, computational power and training time are significant factors in this decision, and I find it is appropriate to focus on a smaller subset of stocks. It may still be a valid possibility to use a wider array either as a robustness check or if constraints allow, but I will leave those possibilities to the future.

In addition to the group of related stocks, a set of macroeconomic variables will be provided as well. Indicator variables for the COVID-19 pandemic and Great Recession will also be provided to the model, though they may be removed as a robustness check. 

There are a few possible ways in which to split the data between training and testing. First, train the model on a portion of the non-COVID-19 data and test on both the remaining portion of that portion and the COVID-19 period data. Second, the model could be trained on random portions of the data regardless of pandemic status. Both methods would provide the ability to compare the model’s performance during both non-pandemic and pandemic test periods, but the appropriateness of including or not-including pandemic data will be left up to future considerations.

An additional finding of Fisher & Krauss (2017) is that the accuracy of their algorithms decreased over time, coinciding with the broad-scale availability of similar algorithms to financial investors. In years prior to 2001, the models displayed a much greater ability to generate profit and accurately determine price changes than in the 2001-2009 period (which in turn was also more predictable than the 2010 – 2015 period). Indeed, the average daily returns for both the LSTM and random forest models were negative for more than half the final period. While this trend may have (and likely did) continued beyond 2015, the volatility caused by the COVID-19 pandemic may provide the opportunity for the random forest model to regain some of its predictive power (indeed, that will be the focus of this paper).

```python
def corr_matrix_heatmap(data, dim = 20):
        fig, ax = plt.subplots(figsize = (dim,dim))
        im = ax.imshow(data, norm = plt.cm.colors.Normalize(-1,1), cmap = "inferno")  
        labels = data.keys()
        num_vars = len(labels)
        tick_labels = [label.replace(" ","\n") for label in labels]
        tick_font_size = 120 / num_vars
        val_font_size = 200 / num_vars
        plt.rcParams.update({'font.size': tick_font_size})
        xticks = np.arange(num_vars)
        plt.xticks(xticks, tick_labels, fontsize = tick_font_size, rotation=90)
        yticks = np.arange(num_vars)
        plt.yticks(yticks, tick_labels, fontsize = tick_font_size)
        # put all options choses above into use and create figure
        for i in range(len(labels)):
            for j in range(len(labels)):
                text = ax.text(i,j, str(round(data.values[i][j], 2)),
                              fontsize = val_font_size, ha = "center",
                               va = "center", color = "w")
        title_font = {"fontname":"Times New Roman"}  
        plt.title("Correlation", fontsize = 50, **title_font)   
        cbar = fig.colorbar(im)
        plt.show()
```

```python
data = pd.read_csv("DATA.csv")
data.head()
```

```python
stock_data = data[["MGM", "LVS", "PENN", "WYNN"]]
corr_data = stock_data.corr()
corr_matrix_heatmap(corr_data, dim = 15)
```

```python
macro_data = data[["UNEM", "CPI", "GDPCAP", "INT"]]

def color_dim_scatter(data):
    for key1 in data:
        for key2 in data:
            #don't plot a var against itself
            if key1 != key2:
                for key3 in data:
                    #make sure all three are different
                    if key1!= key3 and key2 != key3:
                        fig, ax = plt.subplots(figsize = (20,20))
                        data.plot.scatter(x = key1, 
                                          y = key2,
                                          c = key3,
                                          colormap = "inferno",
                                          s = 30, 
                                          alpha = .5, 
                                          ax = ax)
                        #remove tick lines
                        plt.setp(ax.get_xticklines(), visible = False)  
                        plt.setp(ax.get_yticklines(), visible = False)
                        #show fig
                        plt.show()
                        #close fig
                        plt.close()

color_dim_scatter(macro_data)
```
